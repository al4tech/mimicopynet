# BasicCNN を用いた学習．

BasicCNN は， CQT済みデータを入力に，ピアノロール(hold)を出力するモデルです．計算モデルだけでなく，学習用のデータをセットしたり，学習を行ったりするメソッドが付属しています．

```python
model = mcn.model.BasicCNN(input_cnl=2)
```

```python
model.load_cqt_inout("cqt_data/thedata_raw.npz")
```

```python
sum_score = np.zeros(model.score[0].shape)
for s in model.score:
    sum_score += s
```

```python
plt.imshow(sum_score/len(model.score), origin='lower'); plt.colorbar(); plt.show()
```

```python
sum_attack = np.zeros(model.score[0].shape)
attack = np.zeros(model.score[0].shape)
for s in model.score:
    attack[:,1:] = np.logical_and(s[:,:-1]==0, s[:,1:]==1).astype(np.int32)
    sum_attack += attack
plt.imshow(sum_attack/len(model.score), origin='lower'); plt.colorbar(); plt.show()
```
